```python
# Placeholder para a temperatura do sistema. Em um cenário real, você obteria isso de um sensor.
temperature_input = input("Digite a temperatura atual do sistema em Celsius: ")

# Tenta converter a entrada para um número inteiro
try:
    temperature = int(temperature_input)
except ValueError:
    print("Entrada inválida. Por favor, digite um número inteiro para a temperatura.")
    # Define uma temperatura padrão ou encerra o programa, dependendo da necessidade
    temperature = 0 # Ou raise ValueError("Entrada inválida")

# Verifica a temperatura com base nas novas faixas:
if temperature > 37:
    print(f"ALERTA: Temperatura do sistema ({temperature}°C) indica Febre!")
elif temperature < 35:
    print(f"ALERTA: Temperatura do sistema ({temperature}°C) indica Hipotermia!")
else:
    print(f"Temperatura do sistema ({temperature}°C) está Normal.")
```
